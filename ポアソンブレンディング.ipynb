```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math

# src: 合成したい画像
src = cv2.imread('/content/080000.png')
src = cv2.resize(src, dsize = (30, 30))
# 
# target: 合成先
target = cv2.imread('/content/hitachiota07_20190626161233.jpg')
# img[top : bottom, left : right]
target = cv2.resize(target, dsize = (256, 256))
target = target[150 : 180, 150: 180]

# mask: マスク
mask = np.zeros(src.shape, dtype = np.uint8)
cv2.rectangle(mask, (0, 0), (30, 30), (255, 255, 255),thickness = -1)

# target = np.zeros(src.shape, dtype = np.uint8)
# target = cv2.resize(target, dsize = (256, 256))

# 画像合成
img = cv2.seamlessClone(src, target, mask, (30//2, 30//2), cv2.MIXED_CLONE)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```
